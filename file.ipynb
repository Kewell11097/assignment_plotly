```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_csv('titanic3.csv')
```

```python
df.head()
```

```python
#bar plot showing survivors and those who did not survive by gender

fig_bar = px.bar(
    df, x='sex', y='survived', color='survived',
    title='Survival rate by gender'
)
fig_bar.show()
```

```python
#histogram showing distribution of passenger ages, different colors for survivors and non-survivors

fig_histogram = px.histogram(df, x='age', color='survived',
                             nbins=20, title='Age distribution by survial and non-survival',
                             labels={'age':'Age'})

fig_histogram.show()
```

```python
#violin plot to visualize fare prices by class

fig_violin = px.violin(df, x='pclass', y='fare',
                       points='all', title='Fare for passanger classes',
                       labels={'pclass': 'Passenger class'})
fig_violin.show()
```

```python
#scatter plot for relationship between age and fare prices

fig_scatter = px.scatter(df, x='age', y='fare',
                         title='Age vs Price by Class',
                         color='pclass', symbol='pclass')
fig_scatter.show()
```

```python
#faceting for relationship betwee age, fare and class for survivors and who did not

fig_faceting = px.scatter_matrix(df, dimensions=['age', 'fare', 'pclass'],
                                 title='Age vs Fare vs Class', 
                                 labels={'age': 'Age', 'fare': 'Fare', 'pclass': 'Class'},
                                 color='survived')
fig_faceting.show()
```

```python
#export as HTML files

import plotly.io as pio

pio.write_html(fig_bar, 'bar_plot.html')
pio.write_html(fig_histogram, 'histogram_plot.html')
pio.write_html(fig_violin, 'violin_plot.html')
pio.write_html(fig_scatter, 'fig_scatter_plot.html')
pio.write_html(fig_faceting, 'fig_faceting_plot.html')
```
